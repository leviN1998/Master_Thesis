```python
class SimpleNet(torch.nn.Module):

    def __init__(self):
        super(SimpleNet, self).__init__()
        # Define your network layers here
    


    def forward(self, x):
        # Define the forward pass
        x = self.fc1(x)
        x = torch.relu(x)
        x = self.fc2(x)
        x = torch.relu(x)
        x = self.fc3(x)
        x = torch.relu(x)
        x = self.fc4(x)
        x = torch.tanh(x)
        return x  * 1000
```

```python
model = torch.load("../data/ball_gun_to_spin.pth", weights_only=False)
print(model)
```

# Load all files and get their spin

```python
files = [f for f in os.listdir("/home/lkolmar/Documents/metavision/recordings/new_real_dataset/preprocessed/") if f.endswith(".hdf5")]
recordings = []
for f in files:
    if f[4] == "-":
        spin = -int(f[5])
    else:
        spin = int(f[4])
    if f[-9] == "-":
        sidespin = -int(f[-8])
    else:
        sidespin = int(f[-8])
    index = int(f[-6])
    recordings.append((f, spin, sidespin, index))
```

# Save validation set

```python
labels_2 = []
labels_6 = []
idx = 0

for rec in recordings:
    if rec[3] in [0, 1]:
        setting = torch.tensor([[rec[1], rec[2]]], dtype=torch.float32)
        spin = model(setting)
        spin = spin.detach().numpy()
        rps = np.linalg.norm(spin) / (2 * np.pi)
        topspin = spin[0][2] < 0
        if topspin:
            labels_2.append(0)
            if rps < 50:
                labels_6.append(0)
            elif rps < 95:
                labels_6.append(1)
            else:
                labels_6.append(2)
        else:
            labels_2.append(1)
            if rps < 50:
                labels_6.append(3)
            elif rps < 95:
                labels_6.append(4)
            else:
                labels_6.append(5)

        os.makedirs(f"/home/lkolmar/Documents/metavision/recordings/new_real_dataset/dataset/validate/preprocessed/{str(idx).zfill(5)}/", exist_ok=True)
        shutil.copyfile(f"/home/lkolmar/Documents/metavision/recordings/new_real_dataset/preprocessed/{rec[0]}", f"/home/lkolmar/Documents/metavision/recordings/new_real_dataset/dataset/validate/preprocessed/{str(idx).zfill(5)}/{str(idx).zfill(5)}_roi.hdf5")
        idx += 1

print(len(labels_2), len(labels_6))
df = pd.DataFrame({'index': range(len(labels_2)), 'label': labels_2})
df.to_csv("/home/lkolmar/Documents/metavision/recordings/new_real_dataset/dataset/validate/config/labels_2.csv", index=False)

df6 = pd.DataFrame({'index': range(len(labels_6)), 'label': labels_6})
df6.to_csv("/home/lkolmar/Documents/metavision/recordings/new_real_dataset/dataset/validate/config/labels_6.csv", index=False)
```
