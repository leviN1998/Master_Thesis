```python
# Plot the given 3D points and also draw the sphere they lie on
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 (needed for 3D projection)
import numpy as np

# Given points
points = np.array([
    (-0.7871853989303501,-0.15007415395023616,0.5981779802257738),
    (0.831954820950352,0.49069830978374895,-0.25897170631719546),
    (-0.4380651303116413,0.6340433924864243,0.6372502789715757),
    (-0.28577569315413615,-0.0751439630439845,-0.9553458211665122),
    (0.42457666418053897,-0.8890865192523817,0.17105501313030896),
    (0.3937938198260508,-0.7328713388451765,-0.5548207171385062),
    (0.537297995885336,-0.3162797734079114,-0.7818426750636229),
    (0.8773905017538899,-0.3814586799909237,0.2909900047967959),
    (0.9994501631499049,-0.008825817108151595,-0.03196054336212123),
    (-0.3261761708189827,0.4668342563061622,-0.8219944541959748),
    (-0.7718798191634665,-0.3261563958623314,-0.5457321231211046),
    (-0.877879164370942,-0.47184328362705946,0.08180518601782476),
    (0.3052266058050259,0.8307531602335847,0.4654953338870919),
    (0.7947152474101213,0.5394906604541255,0.27816811969857586),
    (-0.048474913819838546,-0.943268670268489,-0.3284728275216626),
    (-0.9626294256281197,0.27079399448256064,-0.0038989058792789862),
    (0.819611785227634,-0.008596255585579969,-0.5728548035112196),
    (-0.5980263812854965,-0.7741974515249317,0.20732282396995058),
    (-0.41500949748885113,0.9086344972508039,0.04637313230557253),
    (-0.03704122064186678,0.8846041257786046,-0.4648693242501934)
])

x, y, z = points[:,0], points[:,1], points[:,2]

# Estimate the radius from the data (should be ~1 for a unit sphere)
r = np.mean(np.linalg.norm(points, axis=1))

# Create a spherical mesh
u = np.linspace(0, 2*np.pi, 80)
v = np.linspace(0, np.pi, 40)
xs = r * np.outer(np.cos(u), np.sin(v))
ys = r * np.outer(np.sin(u), np.sin(v))
zs = r * np.outer(np.ones_like(u), np.cos(v))

# Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Sphere as a wireframe (keeps points visible)
ax.plot_wireframe(xs, ys, zs, linewidth=0.5, rstride=2, cstride=2)

# Scatter the points
ax.scatter(x, y, z, s=40)

# Equal aspect ratio for x, y, z
max_range = np.array([x.max()-x.min(), y.max()-y.min(), z.max()-z.min()]).max()
Xb = (x.max()+x.min())/2
Yb = (y.max()+y.min())/2
Zb = (z.max()+z.min())/2
ax.set_box_aspect([1, 1, 1])
ax.set_xlim(Xb - max_range/2, Xb + max_range/2)
ax.set_ylim(Yb - max_range/2, Yb + max_range/2)
ax.set_zlim(Zb - max_range/2, Zb + max_range/2)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Punkte auf einer Kugel mit eingezeichneter Kugel')

plt.show()
```
